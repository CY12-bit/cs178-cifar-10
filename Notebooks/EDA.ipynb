## Airplane or Banana: Classification for CIFAR-10 

```python
from cifar10_import import *
import cv2
```

```python
# Import the CIFAR-10 Dataset
(x_train, y_train), (x_test, y_test) = importCifar10()
assert x_train.shape == (50000, 32, 32, 3) # For slot in the 32 x 32 grid, it's an list of 3 numbers for RGB
assert x_test.shape == (10000, 32, 32, 3)
assert y_train.shape == (50000, 1)
assert y_test.shape == (10000, 1)
```

Exploratory Data Analysis

```python
print(" Dataset dimensions: ")
print(f"\tTraining data: {x_train.shape}")
print(f"\tTesting data: {x_test.shape}")
```

```python
# CIFAR-10 classes
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']    
# Create a new figure
plt.figure(figsize=(15,15))

# Loop over the first 64 images
for i in range(64):
    # Create a subplot for each image
    plt.subplot(8, 8, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)

    # Display the image
    c = plt.imshow(x_train[i])

    # Set the label as the title
    plt.title(class_names[y_train[i][0]], fontsize=12)
```

```python
x_train[0].shape
```

```python
class_counts = np.zeros(len(class_names))
for label in y_test[:300]:
    class_counts[label[0]] += 1

# Plot histogram
plt.figure(figsize=(10, 6))
plt.bar(class_names, class_counts, color='skyblue')
plt.title('CIFAR-10 Dataset: Number of Images per Class')
plt.xlabel('Class')
plt.ylabel('Number of Images')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
